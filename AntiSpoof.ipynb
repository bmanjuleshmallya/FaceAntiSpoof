```python
import cv2
import sys
# import dlib
from pandas.core import frame
from skimage.feature import local_binary_pattern
import numpy as np

# hog_detector = dlib.get_frontal_face_detector()
faceCascade = cv2.CascadeClassifier('haarcascade/haarcascade_frontalface_default.xml')
video_capture = cv2.VideoCapture(0)

while True:
    ret, frame = video_capture.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = faceCascade.detectMultiScale(
        gray,
        scaleFactor=1.1,
        minNeighbors=5,
        minSize=(30, 30),
        flags=cv2.CASCADE_SCALE_IMAGE
    )

    from pprint import pprint
    pprint(faces)
    
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 0, 255), 2)
        face=gray[y:y+h, x:x+w]
        lbp = local_binary_pattern(face, 24, 8, method="uniform")
        (hist, _) = np.histogram(lbp.ravel(), bins=np.arange(0, 24+3), range=(0, 24+2))
        hist = [hist/sum(hist)]
        valuePredicted = clfSVM.predict(hist)
        str=""
        if valuePredicted == [0]:
            str="Fake"
        elif valuePredicted == [1]:
            str="Real"

        font = cv2.FONT_HERSHEY_SIMPLEX
        cv2.putText(frame,str,(x,y-20), font, 1, (0,255,0), 2)

    # dets = hog_detector(frame, 1)
    # Using HOG detector
    # for i, d in enumerate(dets):
    #     rect = cv2.rectangle(frame, (d.bottom(), d.top()), (d.left(), d.right()), (255, 0, 0), 1)
    #     font = cv2.FONT_HERSHEY_SIMPLEX
    #     cv2.putText(frame,'name',(d.bottom(),d.top()), font, 1, (255,0,0)) 

    cv2.imshow('Video', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
video_capture.release()
cv2.destroyAllWindows()
```
